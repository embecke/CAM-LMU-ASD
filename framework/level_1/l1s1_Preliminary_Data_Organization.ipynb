```python
import os
import shutil #to move files and folders 
```

```python
"""
downloading files and folders directly from lrz and sending them to the parent folder

night time eeg files, reports, (meditation time eeg files and reports)
"""
#download (and extract) the full folder of the participant that you want to analyse


#1. create the participant main folder with the correct participant name in the correct destination (and move EEG/meditation full folder as is into this main folder; also move start and bdi and other general sheets into the main folder; also move overview excel sheet into the main folder)
downloaded_folder = input("Enter path to the downloaded folder: ")
sub_name = input('Participant name which will be the name of the main folder (e.g: "LMU_Stream_HC_002"): ')
dest = input("destination folder where this folder should be created: ")
main_folder = os.path.join(dest, sub_name)
os.mkdir(main_folder)
if os.path.exists(main_folder):
    print("main participant folder created")
else:
    print("couldn't create main participant folder")

#move meditation eeg folder
eeg_meditation_folder = os.path.join(downloaded_folder, 'EEG/Meditation')
try:
    shutil.move(eeg_meditation_folder, main_folder)
    print("moved meditation eeg folder from downloaded folder to main participant folder")
except:
    print("couldn't move folder")

#move start and BDI questionnaire sheets to the main folder
for file in os.listdir(os.path.join(downloaded_folder, 'App')):
    if 'App_Start' in file or 'StartQ' in file:
        start_file = os.path.join(downloaded_folder, 'App', file)
        try:
            shutil.move(start_file, main_folder)
            print("moved Start questionnaire file in App data from downloaded folder to main participant folder")
        except:
            print("couldn't move file")
    elif 'BDI' in file:
        bdi_file = os.path.join(downloaded_folder, 'App', file)
        try:
            shutil.move(bdi_file, main_folder)
            print("moved BDI questionnaire file in App data from downloaded folder to main participant folder")
        except:
            print("couldn't move file")
    
    elif 'EndQ' in file: #seen in ASD_002 in LMU_Stream_Data on LRZ
        end_file = os.path.join(downloaded_folder, 'App', file)
        try:
            shutil.move(end_file, main_folder)
            print("moved End questionnaire file in App data from downloaded folder to main participant folder")
        except:
            print("couldn't move file")
    elif 'Demographics' in file: #seen in ASD_003 and subsequent participants on LRZ (as of 01.01.2025)
        dem_file = os.path.join(downloaded_folder, 'App', file)
        try:
            shutil.move(dem_file, main_folder)
            print("moved Demographics file in App data from downloaded folder to main participant folder")
        except:
            print("couldn't move file")


#move overview sheet from downloaded folder to main participant folder
for file in os.listdir(downloaded_folder):
    if 'Overview' in file:
        overview_file = os.path.join(downloaded_folder, file)
        try:
                    shutil.move(overview_file, main_folder)
                    print("moved overview file from downloaded folder to main participant folder")
        except:
                    print("couldn't move file")      
            
    else:
        continue
```

```python
#2. create folders for each day.
#2.1. Option 1: if previous night EEG data to be paired with next day daytime data [n1_d(2) format]
#2.1.1 create as many folders of incremental nx_dx+1 format as there are night EEG folders in the Night folder of the participant in LRZ (HIGHLY RECOMMENDED: add in the dates into the folder names (as the code does below). This would make it much easier to handle app data file moving later and also force a lookout when some dates have missing data. Use the overview excel file to do this if needed)
"""
#Check if it is in the format Night10_13112023_14112023_HC_002 or Night10_ASD_003_19092024_20092024 or something else (variations keep popping up). Based on the format of the folder name, enter the indices of the required sub strings (from splitting at "_") starting from 0 to end.
Example: if format: Night10_13112023_14112023_HC_002
a1 = 1, a2 = 0, a3 = 2
"""
a1 = int(input("Enter the index of earlier date (night): ")) #index of earlier date (night)
a2 = int(input("Enter the index of 'Night': ")) #index of 'Night'
a3 = int(input("Enter the index of later date (day): ")) #index of later date (day)

for file in sorted(os.listdir(os.path.join(downloaded_folder, 'EEG/Night'))):
    #check if it is in the format Night10_13112023_14112023_HC_002 or Night10_ASD_003_19092024_20092024 or something else (variations keep popping up)
    split_file = file.split('_')    
    day_night_folder_name = split_file[a1] + '_n' + split_file[a2][5:] + '_' + split_file[a3] + '_d'
    print(day_night_folder_name)
    
    
    #day_night_folder_name = split_file[1] + '_n' + split_file[0][5:] + '_' + split_file[2] + '_d'
    day_night_folder = os.path.join(main_folder, day_night_folder_name)
    os.mkdir(day_night_folder)
    
    #2.1.2 move the files in the subfolders in Night folder to the correct incremental folders matched by night number
    try:
            shutil.move(os.path.join(downloaded_folder, 'EEG/Night', file), day_night_folder) 
            print("moved night eeg files from downloaded folder to required subfolder")
    except Exception as e:
            print(f"Couldn't move night EEG files from {file}: {str(e)}")

#Note: If the files available for the night do not match the number of app files available for the day (as is for HC_003), then only the App files (excel sheets) corresponding to the night files and dates available will be moved. Check participant folder for these inconsistencies. If all files are available for the same days, then it works without any issues
#Note: Inconsistent naming pattern for folders within EEG/Night in original downloaded folder between participants. Check to see if naming format is the same as what is coded for above. If not, adjust accordingly (convention change seen in ASD_003 and other participant files in https://syncandshare.lrz.de/getlink/fiTCXvaGemE1iJoxzBdJjh/ 


#2.1.3 move the app data excel sheets. Match the file names by matching the date in the name to the date in the folder name
#Check style and adjust index accordingly as above
a4 = int(input("Enter the index of the date in the excel file: ")) #eg. if excel file name App_13122023_HC_003.xlsx, a4 = 1 (because 13122023 is the date)
for file in sorted(os.listdir(os.path.join(downloaded_folder, 'App'))):
    fileName = file.split(".")[0]
    for folder in sorted(os.listdir(main_folder)):
        #print(file, folder)
        if folder.endswith("d") and fileName.split('_')[a4] in folder.split('_'):
            print('yes', fileName, folder)
            
            try:
                shutil.move(os.path.join(downloaded_folder, 'App', file), os.path.join(main_folder, folder)) 
                print("moved app file from downloaded folder to required subfolder")
                break
            except Exception as e:
                print(f"Couldn't move app file {file} to {folder}: {str(e)}")           
            
         
```

```python
"""
Creating the necessary subfolders for each night-day folder
names of folders to be used
"""
folder1 = 'empatica'
folder2 = 'saved_figures'

folder11 = 'aggr_p_min'
folder12 = 'avro_files'
folder13 = 'avro2csv'
main_folder = input('enter the full filepath of the parent folder: ') #input participant analysis folder as parent folder

for subfolder in os.listdir(main_folder):
    if not(os.path.exists(os.path.join(main_folder, subfolder, folder1))) and subfolder.endswith('_d'):
        #uncomment the below lines if not executed for the parent folder yet
        dir1 = os.mkdir(os.path.join(main_folder, subfolder, folder1))
        dir2 = os.mkdir(os.path.join(main_folder, subfolder, folder2))

        for folder in os.listdir(os.path.join(main_folder, subfolder)):
            if folder.endswith('empatica'):
                if not(os.path.exists(os.path.join(main_folder, subfolder, folder, folder11))):
                    dir11 = os.mkdir(os.path.join(main_folder, subfolder, folder, folder11))
                if not(os.path.exists(os.path.join(main_folder, subfolder, folder, folder12))):
                    dir12 = os.mkdir(os.path.join(main_folder, subfolder, folder, folder12))
                if not(os.path.exists(os.path.join(main_folder, subfolder, folder, folder13))):
                    dir13 = os.mkdir(os.path.join(main_folder, subfolder, folder, folder13))
```

```python
"""
code block to now send the wristband data to the right subfolders
agg per minutes files -> aggr_p_min subfolder
raw files -> avro_files subfolder
"""
#Note: Just like the App data of the first day (corresponding to which no eeg recording was done at night), also moving first day embrace folder entirely to the first folder in main_folder incase this is needed for any new analyses.

#3.1 on LRZ, the folders are labelled by date. Therefore need to access the embrace folders by matching its date to the day subfolders on system. Go to the right subfolder by date and move all the files in "digital_biomarkers/aggregated_per_minute" to 'aggr_p_min' folder
participant_code = input("enter the code used within all the embrace folders within which all the wristband data is contained (e.g: 05-3YK3M16177): ")
for embrace_folder in os.listdir(os.path.join(downloaded_folder, 'EmbracePlus')):
    efolder = embrace_folder.split('-')
    #print("Embrace folder split:", efolder)

    for day_folder in os.listdir(main_folder):
        if len(day_folder.split('_')) > 2:
            day_folder_split = day_folder.split('_')
            #print("Day folder split:", day_folder_split)
            day1 = False
            if len(day_folder_split[1]) == 3 and day_folder_split[1] == 'n01':
                day1 = True
            elif len(day_folder_split[1]) == 2 and day_folder_split[1] == 'n1':
                day1 = True

            if efolder[0] in day_folder_split[0][-4:] and efolder[1] in day_folder_split[0][2:4] and efolder[2] in day_folder_split[0][0:2] and day_folder_split[3] == 'd' and day1 == True:
                print("First day match found!")
                try:                  
                    shutil.move(os.path.join(downloaded_folder, 'EmbracePlus', embrace_folder), os.path.join(main_folder, day_folder)) 
                    print("moved embrace folder of first date from downloaded folder to required date folder: ", day_folder)                
                except Exception as e:
                    print("couldn't move the first date embrace folder to the required date folder: ", day_folder)
                
            
            # Compare the elements of efolder with day_folder_split
            if efolder[0] in day_folder_split[2][-4:] and efolder[1] in day_folder_split[2][2:4] and efolder[2] in day_folder_split[2][0:2] and day_folder_split[3] == 'd':
                print('Match found!')
                #print(day_folder)
                #print("Embrace folder split:", efolder)
                #print("Day folder split:", day_folder_split)
                try:
                    for file in os.listdir(os.path.join(downloaded_folder, 'EmbracePlus', embrace_folder, participant_code, 'digital_biomarkers/aggregated_per_minute')):
                        shutil.move(os.path.join(downloaded_folder, 'EmbracePlus', embrace_folder, participant_code, 'digital_biomarkers/aggregated_per_minute', file), os.path.join(main_folder, day_folder, folder1, folder11)) 
                    print("moved aggregated per minute files from downloaded folder to required subfolder - empatica/aggr_p_min for day: ", day_folder)                
                except Exception as e:
                    print("couldn't move aggregated per minute files for day: ", day_folder)
                #3.2 Similarly, move all the raw files in raw_data/v6 to 'avro_files'
                try:
                    for file_1 in os.listdir(os.path.join(downloaded_folder, 'EmbracePlus', embrace_folder, participant_code, 'raw_data/v6')):
                        shutil.move(os.path.join(downloaded_folder, 'EmbracePlus', embrace_folder, participant_code, 'raw_data/v6', file_1), os.path.join(main_folder, day_folder, folder1, folder12)) 
                    print("moved raw files from downloaded folder to required subfolder - empatica/avro_files for day: ", day_folder)                
                except Exception as e:
                    print("couldn't move raw avro files for day: ", day_folder)

            
```
